Получение текстов из документов


```
RPD_texts = pr.extract_text_from_pdfs_recursively('/content/drive/My Drive/Magistracy_SPbU/3_semester/Study_practice/RPD')
```

Преобразование текстов в датафрейм


```
df = pd.DataFrame(RPD_texts, columns=['raw_text'], dtype='str')
df.info()
```

```
df.head()
```

# Очистка текстов

```
df['clean_text'] = df['raw_text'].apply(pr.clean_text)
df['clean_text_no_stopwords'] = df['clean_text'].apply(pr.stop_words_remove)
df.head()
```

```
tfidf = TfidfVectorizer(max_features=50000, ngram_range=(1, 2))
Vectorized_text = tfidf.fit_transform(df.clean_text_no_stopwords)
Vectorized_text
```
